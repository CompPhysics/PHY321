```python
#In this program, we will be evaluating the equation of motion for a single pendulum using Lagrangian
#Mechanics. The differential equations involved are solved using the Euler-Cromer method and sympy documentation.

# import methods from sympy
from sympy import *                

#coordinates
x, t = symbols('x t')         
m,l=symbols(' m l')

## import the Cartesian coordinate system class
from sympy.vector import CoordSys3D, Del
## construct a coordinate system called 'R'
R = CoordSys3D('R')                        

#Set parameters
g=9.8

x=Function('x')(t)
v=diff(x, t)

#Define kinetic and potential energies 
KE=0.5*m*l**2*(v)**2
PE=m*g*l*(1-cos(x))

#Define Lagrangian
L=KE-PE

#use Euler-Lagrange Equation
LHS=diff(L, x)
RHS=diff(L, v)
RHS=diff(RHS, t)
RHS, LHS

diffeq=RHS+LHS
diffeq
simplify(diffeq)
#setting this to zero gives the diffeq that describes the motion of the pendulum
```

```python
import numpy as np
import matplotlib.pyplot as plt
## Euler-Cromer Integration

## Set parameters
m=1
l=1
g=9.8
dt=0.001
time = np.arange(0,10,dt)

## Initialize position, velocity, and acceleration
pos = np.zeros(len(time))
vel = np.zeros(len(time))
accel = np.zeros(len(time))

## Set the initial position and velocity
pos[0] = np.pi/3
vel[0] = 0

## Set up the counter
i = 0   

## Calculation loop
while i < len(time)-1:
    
    ## Perform Euler-Cromer integation for the exact differential equation
    x_dub1=g/l*np.sin(pos[i])
    accel[i]=x_dub1
    
    vf1=vel[i]+x_dub1*dt
    vel[i+1]=vf1
    
    xf1=pos[i]+vf1*dt
    pos[i+1]=xf1
    
    i+=1
    
## Plot results
plt.plot(time, pos, color='green', label="x(t)") ## Position
plt.plot(time, vel, color='blue', label='v(t)') ##Velocity 
plt.plot(time, accel, color='red', label='v(t)') ##
plt.xlim([0,2*np.pi])
plt.title('Position, Velocity, and Acceleration vs. Time')
plt.xlabel("time (s)")
plt.ylabel("Solutions")
plt.legend()
plt.show()
```
