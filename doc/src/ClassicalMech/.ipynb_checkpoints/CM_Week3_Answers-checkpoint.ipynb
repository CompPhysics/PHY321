#### Let's begin with a simple rocket problem

Suppose I have a rocket starting with mass $M_0$ traveling upwards against gravity. As it is propeled upwards by an exaust velocity of $v_e$ the rocket loses mass from used fuel at a rate of $\dfrac{dM}{dt}$. 

##### How can we computationally solve for the velocity of the rocket at time $t$?
Well that's a great question, and we can use simple computational analysis utilizing momentum to solve for this.


#### Note: The exaust velocity is NOT the velocity of the rocket


#### Let's begin by analyzing our problem first:
- We know the final mass of therocket must be $M_0 - M_{fuel}$
- momentum, $p = mv$, must be conserved, thus:
- $p_i = p_1 + p_2$
- We're always losing mass, so we can analyze the problem with 

#### Given these intial conditions, use Euler's Method to solve for this problem:
- $M_0 = 2000 kg$
- $v_{rocket_{initial}} = 0 kg$
- $v_e = 100 m/s$
- $M_{fuel} = 1400 kg$
- $\dfrac{dM}{dt} = 50 kg/s$

```python
# Let's begin by importing our basic packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Now let's initialize the values we do know

M_0 = 2000 #kg
M_f = 600 #kg
M = np.arange(M_f,M_0+1)
l = len(M)
```

```python
d = M[l:(0):-1]
```

```python
v=[0]
for i in range(0, len(d)):
    v.append((d[i]*v[i]-100)/M[i])
```
